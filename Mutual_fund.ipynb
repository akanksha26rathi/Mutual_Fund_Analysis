# Dasboard

```python
print(df.category.unique())
f = input("Enter category name:")
cat = df[df.category==f]

print(cat.risk_type.unique())
r = input("Risk type:")
risk = cat[cat.risk_type==r]

print(risk.AMC_name.unique())
n = input("AMC Name")
name = risk[risk.AMC_name==n]

y=name.groupby('Mutual_Fund_Name')['return_1yr'].mean()
sb.barplot(y , color='brown')
plt.xticks(rotation=90)
plt.show()
```

```python
z=name.groupby('Mutual_Fund_Name')['return_3yr'].mean()
sb.barplot(z , color='brown')
plt.xticks(rotation=90)
plt.show()
```

```python
x=name.groupby('Mutual_Fund_Name')['return_5yr'].mean()
sb.barplot(x ,  color='brown')
plt.xticks(rotation=90)
plt.show()
```

# Subplot

```python
fig = plt.figure(figsize=[20,10])

fig.add_subplot(221)
sb.barplot(z , color='magenta')
plt.xticks(rotation=90)

fig.add_subplot(222)
sb.barplot(y , color='magenta')
plt.xticks(rotation=90)

fig.add_subplot(223)
sb.barplot(x, color='magenta')
plt.xticks(rotation=90)
plt.show()
```
